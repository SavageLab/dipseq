```python
import glob
import pylab
import numpy as np

from Bio import SearchIO
from utils.read_matches import ReadMatches

# Filenames for the first end of the paired ends for the A group (for now).
insert_aligned_fnames = glob.glob('../_read_filter_data/dana-A*R1*_insert_aligned.psl')
backbone_aligned_fnames = glob.glob('../_read_filter_data/dana-A*R1*_backbone_aligned.psl')
insert_aligned_fnames = sorted(insert_aligned_fnames)
backbone_aligned_fnames = sorted(backbone_aligned_fnames)
print insert_aligned_fnames
print backbone_aligned_fnames
```

```python
# IDs of reads that don't make sense.
suspect_ids = set()

# Maps read ID to ReadMatches object.
all_matches = {}

# Figure out where the insert (GFP) inserted in the backbone (MBP)
for insert_aligned_fname in insert_aligned_fnames:
    insert_aligned_reader = SearchIO.parse(insert_aligned_fname, 'blat-psl')
    for record in insert_aligned_reader:
        for hsp in record.hsps:
            query_id = hsp.query_id
            
            if (query_id in all_matches):
                # Right now we are making suspect any read that 
                # matched > 1 time. Not sure this makes sense...
                # TODO: what if the matches are on different strands
                # or to different ends of GFP. Should we st
                suspect_ids.add(query_id)

            # Store the location of the range in the query that matches the insert.
            matches = ReadMatches(query_id)
            matches.insert_match = hsp
            all_matches[query_id] = matches
```

```python
# Remove matches with query IDs that were suspect.
# For now this means that we got 2 GFP hits in the same read.
print len(suspect_ids), 'suspect IDs'
total_matches = len(suspect_ids) + len(all_matches)
print 'Of', total_matches, 'total matches'
print '%.2f%% ignored' % (100.0*(float(len(suspect_ids)) / total_matches))
for suspect_id in suspect_ids:
    if suspect_id in all_matches:
        del all_matches[suspect_id]
```

```python
backbone_aligned_ids = set()
hit_counts = {}
distances_from_insert = {}

# Figure out where in the read the backbone (MBP) was found.
for backbone_aligned_fname in backbone_aligned_fnames:
    backbone_aligned_reader = SearchIO.parse(backbone_aligned_fname, 'blat-psl') 
    for record in backbone_aligned_reader:
        for hsp in record.hsps:
            query_id = hsp.query_id
            query_strand = hsp.fragments[0].query_strand
            
            if query_id not in all_matches:
                # We may have deleted the read ID if we considered it suspect
                continue

            # Store the range in the query that matches  the backbone
            matches = all_matches.get(query_id)
            matches.backbone_match = hsp
```

```python
read_ids_to_ignore = set()
n_consistent = 0
n_total = len(all_matches)
for matches in all_matches.values():
    # An incomplete match is also inconsistent.
    if matches.Consistent():
        n_consistent += 1
    else:
        read_ids_to_ignore.add(matches.read_id)

frac_inconsistent = 1.0 - float(n_consistent) / n_total
print '%.2f%% of matches are inconsistent' % (100.0*frac_inconsistent)
```

```python
n_3p = sum([1 for m in all_matches.itervalues() if m.insert_match_end == '3p'])
n_5p = sum([1 for m in all_matches.itervalues() if m.insert_match_end == '5p'])
total = len(all_matches)
frac_3p = float(n_3p) / float(total)
frac_5p = float(n_5p) / float(total)
pct_3p = 100.0 * frac_3p
pct_5p = 100.0 * frac_5p
print '%.2f%% are 3prime matches' % pct_3p
print '%.2f%% are 5prime matches' % pct_5p
print 'sum is %.2f%%' % (pct_3p + pct_5p)
```

```python
print 'Removing', len(read_ids_to_ignore), 'read IDs.'
for read_id in read_ids_to_ignore:
    del all_matches[read_id]
```

```python
pylab.figure(figsize=(20,20))
sorted_read_ids = sorted(all_matches.keys())
y = 1

# Plot the insert and backbone matches along the linear read
# for a sample of reads.
for read_id in sorted_read_ids:
    matches = all_matches[read_id]
    ys = [y, y]
        
    insert_start, insert_end = matches.insert_match_range
    pylab.plot(matches.insert_match_range, ys, 'b--', lw=3)
    pylab.plot([insert_start], [y], 'bo')
    pylab.plot([insert_end], [y], 'b^')
    
    backbone_start, backbone_end = matches.backbone_match_range
    pylab.plot(matches.backbone_match_range, ys, 'g-', lw=3)
    pylab.plot([backbone_start], [y], 'go')
    pylab.plot([backbone_end], [y], 'g^')
    
    if y > 150:
        break
    y += 3
pylab.show()
    
```

```python
hit_counts = {}
for read_id, matches in all_matches.iteritems():
    location = matches.match_location_in_backbone
    hit_counts[location] = hit_counts.get(location, 0) + 1

pylab.figure(figsize=(11,11))
positions = np.array(sorted(hit_counts.keys()))
counts = np.array([hit_counts[k] for k in positions])
mean_count = np.mean(counts) 
stddev = np.std(counts)

total_count = np.sum(counts)
threshold = mean_count / 2.0
idx_above_thresh = np.where(counts > threshold)
my_counts = counts[idx_above_thresh]
my_positions = positions[idx_above_thresh]

pylab.yscale('log')
pylab.plot(positions, counts, 'r.')
pylab.plot(positions, mean_count * np.ones(positions.size),
           'b-', label='mean')
pylab.plot(positions, threshold * np.ones(positions.size),
           'g-', label='threshold')

for position, count in zip(my_positions, my_counts):
    pylab.text(position, count, '%s (%.1f)' % (position, position / 3.0))

pylab.xlabel('Insert position')
pylab.ylabel('Count')
pylab.legend()
pylab.show()
```

```python
print 'Total count', total_count
print 'Mean count %.3f +/- %.3f ' % (mean_count, stddev)
print 'Count threshold', threshold
print
print 'Position => count:'
for position, count in zip(my_positions, my_counts):
    pct_of_total = (float(count) / float(total_count)) * 100.0
    print '\t', position, '=>', count, '(%.2f%% of total, AA %.2f)' % (pct_of_total, position / 3.0)
```
